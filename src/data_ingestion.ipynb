#### Step 1: Fetch Movie Data from API

```python
import pandas as pd
import requests
import json
```

```python
with open("../config/config.json") as config_file: #to load  the config file 
    config = json.load(config_file)

api_token = config["TMDB_API_KEY"] # retrieves the api key from the config file

headers = {
    "Authorization": f"Bearer {api_token}" # sets the authorization header with the api key
}

movie_ids = [0, 299534, 19995, 140607, 299536, 597, 135397,
            420818, 24428, 168259, 99861, 284054, 12445,
            181808, 330457, 351286, 109445, 321612, 260513]

movies_data = []

for movie_id in movie_ids:
    # API endpoint
    url = f"https://api.themoviedb.org/3/movie/{movie_id}"
    
    #sends GET request to the API endpoint with the authorization header for the movie with the given ID 
    response = requests.get(url, headers=headers)
    
    # Check if request was successful
    if response.status_code == 200:
        movies_data.append(response.json())
        print(f"Successfully fetched movie ID: {movie_id}")
    else:
        print(f"Failed to fetch movie ID: {movie_id}, Status code: {response.status_code}")

# Convert the list of movie data to a DataFrame
raw_movies_df = pd.DataFrame(movies_data)

# Save to CSV
raw_movies_df.to_csv("../data/raw/raw_tmdb_movies.csv", index=False)
```

## Fetching Reviews By Movie Id List

```python
import pandas as pd
import requests
import json

# Load API token from config file
with open("../config/config.json") as config_file:  # to load the config file
    config = json.load(config_file)

api_token = config["TMDB_API_KEY"]  # retrieves the API key from the config file

headers = {
    "Authorization": f"Bearer {api_token}"  # sets the authorization header with the API key
}

movie_ids = [
    0, 299534, 19995, 140607, 299536, 597, 135397,
    420818, 24428, 168259, 99861, 284054, 12445,
    181808, 330457, 351286, 109445, 321612, 260513
]

reviews_data = []

for movie_id in movie_ids:
    # API endpoint for reviews
    url = f"https://api.themoviedb.org/3/movie/{movie_id}/reviews"
    
    # Sends GET request to the API endpoint with the authorization header
    response = requests.get(url, headers=headers)
    
    # Check if request was successful
    if response.status_code == 200:
        reviews = response.json().get("results", [])
        for review in reviews:
            review["movie_id"] = movie_id  # Add movie_id to the review data for reference
            reviews_data.append(review)
        print(f"Successfully fetched reviews for movie ID: {movie_id}")
    else:
        print(f"Failed to fetch reviews for movie ID: {movie_id}, Status code: {response.status_code}")

# Convert the list of reviews data to a DataFrame
reviews_df = pd.DataFrame(reviews_data)

# Save to CSV
reviews_df.to_csv("../data/raw/tmdb_movie_reviews.csv", index=False)

print("Reviews data saved to tmdb_movie_reviews.csv")
```

```python
def eval_json(df, separator="|", key=None, column_key_map=None):
   
    default_key = key if key is not None else 'name'
   
    if column_key_map is None:
        column_key_map = {}
  
    result_df = df.copy()
   
    list_dict_columns = []
    single_dict_columns = []
   
    sample_size = min(1000, len(df))
    sample_df = df.head(sample_size)
    
    for column in df.columns:
        has_list_dicts = False
        has_single_dicts = False
        
        for val in sample_df[column].dropna():
            if isinstance(val, list) and val and all(isinstance(i, dict) for i in val if i is not None):
                has_list_dicts = True
                break
            elif isinstance(val, dict):
                has_single_dicts = True
                break
       
        if has_list_dicts:
            list_dict_columns.append(column)
        elif has_single_dicts:
            single_dict_columns.append(column)
    
    for column in list_dict_columns:
        column_key = column_key_map.get(column, default_key)
        def extract_values(x):
            if isinstance(x, list) and x:
                return separator.join(item.get(column_key, '') for item in x if isinstance(item, dict))
            return None
        result_df[column] = df[column].map(extract_values, na_action='ignore')
    
    for column in single_dict_columns:
        column_key = column_key_map.get(column, default_key)
        result_df[column] = df[column].map(
            lambda x: x.get(column_key, '') if isinstance(x, dict) else None,
            na_action='ignore'
        )
    return result_df
```
